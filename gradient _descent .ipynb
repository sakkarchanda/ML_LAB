```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv("advertising.csv")
X = data['TV'].values
Y = data['Sales'].values
X = (X - X.mean()) / X.std()
m = 0
b = 0

# Hyperparameters
L = 0.01
epochs = 1000
n = len(X)
for i in range(epochs):
    Y_pred = m * X + b

    Dm = (-2/n) * np.sum(X * (Y - Y_pred))
    Db = (-2/n) * np.sum(Y - Y_pred)

    m = m - L * Dm
    b = b - L * Db
print("Slope (m):", m)
print("Intercept (b):", b)

plt.scatter(X, Y)
plt.plot(X, m * X + b)
plt.xlabel("Normalized TV")
plt.ylabel("Sales")
plt.title("Gradient Descent Linear Regression")
plt.show()
```
